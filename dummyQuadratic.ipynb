```python
import json
from qiskit_optimization import QuadraticProgram
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit_algorithms import NumPyMinimumEigensolver

# 1. Initialize your WLS/Chebyshev problem (Dummy QUBO for example)
qp = QuadraticProgram()
qp.binary_var('q0')
qp.binary_var('q1')
qp.binary_var('q2')

# Example Objective: Minimize -2*q0 - q1 + q0*q1 + q1*q2
qp.minimize(linear={'q0': -2, 'q1': -1}, quadratic={('q0', 'q1'): 1, ('q1', 'q2'): 1})

# 2. Initialize the exact classical eigensolver
exact_solver = NumPyMinimumEigensolver()

# 3. Wrap it in the MinimumEigenOptimizer
exact_optimizer = MinimumEigenOptimizer(exact_solver)

# 4. Solve the QuadraticProgram
result = exact_optimizer.solve(qp)

# 5. Extract the important data points
optimal_bits = result.x       # Returns a NumPy array of the optimal bitstring e.g., [1., 1., 0.]
optimal_cost = result.fval    # Returns the objective function value (your minimum error)
opt_status = result.status    # Returns an Enum indicating success/failure

# 6. Store the results in a dictionary for easy saving
saved_results = {
    "optimal_bitstring": [int(bit) for bit in optimal_bits], # Convert floats to ints for clean JSON
    "minimum_error": optimal_cost,
    "status": opt_status.name
}

print("Optimization Results:")
print(json.dumps(saved_results, indent=4))

# (Optional) Write to a file
# with open('classical_baseline_results.json', 'w') as f:
```
